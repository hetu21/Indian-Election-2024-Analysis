## In 154 seats BJP came second 

How many seats did bjp lost by equal to less than 10,000 margin 
Pseudo code : 
    Where second winner is BJP/NDA:
    minus the First and second total votes to find margin, then add margin column and then find margin that is more than 15,000

```python
bjp_lost = df_copy[df_copy["second_winner"] == "Bharatiya Janata Party"]
margin_list = []
for consti in bjp_lost["Constituency"].unique():
    dummy = bjp_lost[bjp_lost["Constituency"] == consti].reset_index()
    margin = dummy["Total Votes"][0] -  dummy["Total Votes"][1]
#    bjp_lost["Margin"] = margi
    margin_list.append(margin)
```

```python
margin_series = pd.Series(margin_list)
margin_series[margin_series < 30000].count()
```

Vote Share Analysis - 

```python
#Vote share for one state
gujarat_df = df[df["State"] == "Gujarat"]
grouped_df = gujarat_df.groupby("Party Name")["Total Votes"].sum().sort_values(ascending = False).reset_index().head(5)
grouped_df
```

```python
#voteshare = total votes of the party / total votes polled 
total_polled_votes = gujarat_df["Total Votes"].sum()
grouped_df["Vote Share"] = (grouped_df["Total Votes"]/total_polled_votes)*100
grouped_df
```

```python
states_list = df["State"].unique().tolist()
```

```python
vote_share_df = pd.DataFrame()
for state in states_list:
    state_df = df[df["State"] == state]
    state_grouped = state_df.groupby("Party Name")["Total Votes"].sum().sort_values(ascending = False).reset_index().head(4)
    total_votes_state = state_df["Total Votes"].sum()
    state_grouped["Vote Share"] = (state_grouped["Total Votes"]/total_votes_state)*100
    state_grouped["State"] = state
    vote_share_df = pd.concat([vote_share_df, state_grouped], ignore_index=True)

    
    
```

```python
import matplotlib.pyplot as plt

# Loop through each state and plot the bar chart
for state in vote_share_df['State'].unique():
    # Filter the DataFrame for the current state
    state_data = vote_share_df[vote_share_df['State'] == state]
    
    # Plot the bar chart
    plt.figure(figsize=(10, 6))
    plt.bar(state_data['Party Name'], state_data['Vote Share'], color='skyblue')
    
    # Add titles and labels
    plt.title(f'Vote Share by Party in {state}', fontsize=14)
    plt.xlabel('Party Name', fontsize=12)
    plt.ylabel('Vote Share (%)', fontsize=12)
    
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')
    
    # Show the plot
    plt.tight_layout()
    plt.show()
```
